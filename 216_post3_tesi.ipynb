# **Post Test 3**
Minimumkan:

`f(x, y)` = x<sup>2</sup> + y<sup>2</sup>
dengan batasan: `−5.12 ≤ x, y ≤ 5.12`.
Temukan solusi minimumnya!

```python
import math
import random

def objective_function(x, y):
    """Fungsi tujuan: f(x, y) = x² + y²"""
    return x**2 + y**2

def simulated_annealing():
    # Parameter Simulated Annealing
    initial_temp = 1000
    final_temp = 1e-8
    alpha = 0.9  # Cooling rate
    max_iter = 1000  # Iterasi per suhu

    # Batasan
    x_min, x_max = -5.12, 5.12
    y_min, y_max = -5.12, 5.12

    # Solusi awal (random dalam batasan)
    current_x = random.uniform(x_min, x_max)
    current_y = random.uniform(y_min, y_max)
    current_energy = objective_function(current_x, current_y)

    # Solusi terbaik
    best_x, best_y = current_x, current_y
    best_energy = current_energy

    # Simulated Annealing process
    temperature = initial_temp
    iteration = 0

    print("=== SIMULATED ANNEALING OPTIMIZATION ===")
    print(f"Solusi awal: x = {current_x:.6f}, y = {current_y:.6f}, f(x,y) = {current_energy:.6f}")
    print("-" * 60)

    while temperature > final_temp:
        for i in range(max_iter):
            # Generate neighbor solution
            neighbor_x = current_x + random.uniform(-0.5, 0.5)
            neighbor_y = current_y + random.uniform(-0.5, 0.5)

            # Pastikan dalam batasan
            neighbor_x = max(x_min, min(x_max, neighbor_x))
            neighbor_y = max(y_min, min(y_max, neighbor_y))

            neighbor_energy = objective_function(neighbor_x, neighbor_y)

            # Hitung perubahan energi
            energy_delta = neighbor_energy - current_energy

            # Decision: terima solusi baru atau tidak
            if energy_delta < 0:
                # Jika lebih baik, selalu terima
                current_x, current_y = neighbor_x, neighbor_y
                current_energy = neighbor_energy

                # Update solusi terbaik
                if neighbor_energy < best_energy:
                    best_x, best_y = neighbor_x, neighbor_y
                    best_energy = neighbor_energy
            else:
                # Jika lebih buruk, terima dengan probabilitas tertentu
                probability = math.exp(-energy_delta / temperature)
                if random.random() < probability:
                    current_x, current_y = neighbor_x, neighbor_y
                    current_energy = neighbor_energy

        # Cooling
        temperature *= alpha
        iteration += 1

        # Print progress setiap 50 iterasi
        if iteration % 50 == 0:
            print(f"Iterasi {iteration:3d}: Temp = {temperature:.6f}, Best f(x,y) = {best_energy:.10f}")

    print("-" * 60)
    print("=== HASIL OPTIMASI ===")
    print(f"Solusi minimum ditemukan:")
    print(f"x = {best_x:.10f}")
    print(f"y = {best_y:.10f}")
    print(f"f(x,y) = {best_energy:.10f}")
    print(f"Jumlah iterasi: {iteration}")

    return best_x, best_y, best_energy

# Jalankan optimasi
if __name__ == "__main__":
    # Lakukan beberapa run untuk memastikan konsistensi
    print("Menjalankan Simulated Annealing...")
    print()

    results = []
    for run in range(3):
        print(f"RUN {run + 1}:")
        x_opt, y_opt, f_opt = simulated_annealing()
        results.append((x_opt, y_opt, f_opt))
        print()

    # Tampilkan hasil terbaik dari semua run
    best_run = min(results, key=lambda x: x[2])
    print("=" * 60)
    print("HASIL TERBAIK DARI SEMUA RUN:")
    print(f"x = {best_run[0]:.10f}")
    print(f"y = {best_run[1]:.10f}")
    print(f"f(x,y) = {best_run[2]:.10f}")
    print("=" * 60)

    # Verifikasi solusi teoritis
    print("\nVERIFIKASI:")
    print("Solusi teoritis seharusnya: x = 0, y = 0, f(x,y) = 0")
    print(f"Error: {abs(best_run[2] - 0):.10f}")
```
